Una variable es un nombre que guarda un valor. Se puede usar ese nombre después.

```python
nombre = "Ibuprofeno"
print(nombre)
```

El texto también es un valor. Se puede guardar en una variable y medir.

```python
texto = "Ghosting, mansplaining y scrollear son anglicismos de uso frecuente."
print(texto)
print(len(texto))
```

Y sobre ese texto se puede aplicar un patrón regex — el mismo que se usó en regexr.com.

```python
import re

resultado = re.findall(r"\b\w+ing\b", texto)
print(resultado)
```
