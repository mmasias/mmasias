## El reto pendiente: emojis

En regexr.com y en Google Sheets el patrón para emojis no funcionaba: ambas herramientas usan motores que no manejan bien el rango Unicode completo. Python no tiene esa limitación.

El corpus tiene varios emojis. El bloque siguiente los extrae todos.

```python
patron_emoji = '[\U0001F300-\U0001FAFF]'

emojis = re.findall(patron_emoji, corpus)
frecuencias_emoji = Counter(emojis)
print(frecuencias_emoji.most_common())
```

Nótese que el patrón usa una cadena normal, no una cadena raw (`r"..."`). Con `r'...'`, Python no interpreta `\U` como código Unicode y el patrón no funciona. Es uno de los pocos casos en que la `r` delante de las comillas no se puede usar.

Los bloques Unicode permiten clasificar los emojis por categoría semántica. El bloque siguiente asigna cada emoji a un grupo según su rango de código.

```python
categorias = {
    'naturaleza / lugares': '[\U0001F300-\U0001F5FF]',
    'transporte / objetos': '[\U0001F680-\U0001F6FF]',
    'objetos / comunicacion': '[\U0001F4E0-\U0001F4FF]',
    'personas / gestos':     '[\U0001F900-\U0001FAFF]',
}

for nombre, patron in categorias.items():
    encontrados = re.findall(patron, corpus)
    if encontrados:
        print(f'{nombre}: {encontrados}')
```
