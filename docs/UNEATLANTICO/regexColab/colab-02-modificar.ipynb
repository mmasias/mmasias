## Tarea 1 — Anglicismos con otro sufijo

El patrón actual encuentra palabras terminadas en *-ing*. Modificarlo para que encuentre palabras terminadas en *-er*.

```python
patron = r"\b\w+ing\b"  # ← modifica esta línea

resultados = re.findall(patron, corpus)
print(resultados)
print(len(resultados))
```

## Tarea 2 — De hashtags a menciones

El patrón actual encuentra hashtags (`#lenguaje`, `#corpus`…). Modificarlo para que encuentre menciones en su lugar.

```python
patron = r"#\w+"  # ← modifica esta línea

resultados = re.findall(patron, corpus)
print(resultados)
print(len(resultados))
```

## Tarea 3 — Un carácter, más coincidencias

El patrón actual encuentra URLs que empiezan por `https://`. Modificarlo para capturar también las que empiezan por `http://`.

```python
patron = r"https://\S+"  # ← modifica esta línea

resultados = re.findall(patron, corpus)
print(resultados)
print(len(resultados))
```
