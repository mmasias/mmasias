## El mismo patrón, otra herramienta

En regexr.com, el patrón `\b\w+ing\b` encontraba palabras terminadas en *-ing*. En Google Sheets, `REGEXEXTRACT` devolvía la primera coincidencia de cada celda.

En Python, `re.findall()` devuelve **todas** las coincidencias del corpus completo de una sola vez.

```python
import re
```

```python
anglicismos = re.findall(r"\b\w+ing\b", corpus)
print(anglicismos)
```

¿Qué captura el patrón además de anglicismos?

```python
print(len(anglicismos))
```

## Otro patrón: términos entre comillas

En Sheets, `REGEXEXTRACT` con `"([^"]+)"` devolvía el primer término entre comillas de cada celda. Aquí, sobre el corpus completo de una sola vez:

```python
entre_comillas = re.findall(r'"([^"]+)"', corpus)
print(entre_comillas)
```

```python
print(len(entre_comillas))
```
